```python
import pickle
import requests
import pandas as pd
import matplotlib.pyplot as plt 
with open('dict.pkl','rb') as f:
    artist_dict = pickle.load(f)
```

```python
artist_df = pd.DataFrame(artist_dict).T
```

```python
artist_df.reset_index(inplace=True)
```

```python
sorted_artist_df = artist_df.sort_values('Total Number of Plays',ascending = False)
```

```python
plt.figure(figsize = (12,8));
artist_df.sort_values('Total Number of Plays',ascending = False,inplace=True)
artist_df.head(10).sort_values('Total Number of Plays').plot(kind='barh',y='Total Number of Plays',x='index',color='magenta')
plt.title("Horizontal Bar Chart of Artists with Most Most Plays in Sample Data",fontsize = 18)
plt.xlabel('Total Number of Plays')
plt.ylabel('Artist')
plt.savefig('./images/barh1_mostplays.png', bbox_inches='tight')
```

```python
plt.figure(figsize = (12,9))
artist_df.sort_values('Average Number of Plays',ascending = False,inplace=True)
artist_df.head(10).sort_values('Average Number of Plays').plot(kind='barh',y='Average Number of Plays',x='index',color='orange')
plt.title("Horizontal Bar Chart of Artists with Highest Average Plays in Sample Data",fontsize = 18)
plt.xlabel('Average Number of Plays')
plt.ylabel('Artist')
plt.savefig('./images/barh2_avgplays.png', bbox_inches='tight')

```

```python
plt.figure(figsize = (12,6))
artist_df.sort_values('Number of ratings',ascending = False,inplace=True)
artist_df.head(10).sort_values('Number of ratings').plot(kind='barh',y='Number of ratings',x='index')
plt.title("Horizontal Bar Chart of Most Rated Artists in Sample Data",fontsize = 18)
plt.xlabel('Total Number of Ratings')
plt.ylabel('Artist')
plt.savefig('./images/barh3_nratings.png', bbox_inches='tight')
```

```python
artist_df['Total Number of Plays'].hist(bins = 2000,color = 'magenta')
plt.xlim(0,6000);
plt.title('Total Number of Plays vs Frequency',fontsize = 18)
plt.savefig('./images/hist1_nplays.png', bbox_inches='tight')
```

```python
artist_df['Average Number of Plays'].hist(bins = 1000,color = 'orange')
plt.xlim(0,1500);
plt.title('Average Number of Plays vs Frequency',fontsize = 18)
plt.savefig('./images/hist2_avgplays.png', bbox_inches='tight')
```

```python
artist_df['Number of ratings'].hist(bins = 2000,color = 'steelblue')
plt.xlim(0,50)
plt.title('Number of Ratings vs Frequency')
plt.savefig('./images/hist3_nratings.png', bbox_inches='tight')
```
